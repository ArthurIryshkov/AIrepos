### Задание 1.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
ratings.head()
```

```python
def rate(num):
    if num <= 2:
        return 'низкий рейтинг'
    elif num <= 4:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
```

```python
movies = pd.read_csv('movies.csv')
```

```python
joined = movies.merge(ratings, on='movieId', how='left')
```

```python
rat_mean = joined.groupby('title').mean('rating')
```

```python
rat_mean['class'] = rat_mean['rating'].apply(rate)
```

```python
rat_mean[['rating', 'class']].sort_values('rating', ascending=False).head(20)
```

### Задача 2.

```python
keys = pd.read_csv('keywords.csv')
```

```python
keys
```

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
def geoclass (word):
    for k, v in geo_data.items():
        if word in v:
            return k
    else:
        return 'undefined'
```

```python
keys['region'] = keys['keyword'].apply(geoclass)
```

```python
keys
```

```python
only_reg = keys[keys['region'] != 'undefined']
```
