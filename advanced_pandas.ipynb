### Задание 1.

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
```

```python
def source(row):
    if 'yandex' in row['traffic_source'] or 'google' in row['traffic_source']:
        return 'organic'
    elif ('paid' in row['traffic_source'] or 'email' in row['traffic_source']) and 'Russia' in row['region']:
        return 'ad'
    elif 'paid' in row['traffic_source'] or 'email' in row['traffic_source']:
        return 'other'
    else:
        return row['traffic_source']
```

```python
log['source_type'] =  log.apply(source, axis=1)
```

### Задание 3.

```python
ratings = pd.read_csv('ratings.csv')
```

```python
users = ratings.groupby('userId').agg({'timestamp': ['min', 'max'], 'rating': 'count'})
```

```python
users.loc[users['rating']['count'] > 100]
```

```python
users['lifetime'] = users['timestamp']['max'] - users['timestamp']['min']
```

```python
users.loc[users['rating']['count'] > 100].agg('mean')['lifetime'] / (3600*24) # 'это в днях получается'
```
