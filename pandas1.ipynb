### Задание 2.

```python
power = pd.read_csv('power.csv')
```

```python
power.info()
```

```python
power.head()
```

```python
power2 = power [power['category'].isin([4, 12, 21]) & power['country'].str.contains('Lithuania|Latvia|Estonia')]
```

```python
power_final = power2.query('year < 2010 & year > 2005').query('quantity > -1')
```

```python
power_final.sort_values('quantity', ascending=False)
```

```python
print(sum(power_final['quantity']))
```
