```python
q1 = h3['pulse'].quantile(0.25)
q3 = h3['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = h3[h3['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

```python
pd.concat([h3, remove_outliers]).drop_duplicates(keep=False) # выбросы по пульсу 20 - просто завышенные показатели и NaN.
```

### Участь лошади не влияет на отсутствие показателей температуры и пульса:

### Данные по тем, кому делали операцию, отсутствуют в два раза чаще:

```python
h3[h3['rectal temperature'].isnull()].groupby('surgery')['surgery'].count()
```

```python
h3[h3['pulse'].isnull()].groupby('surgery')['surgery'].count()
```

### В целом, можно считать, что пропуски данных случайны и не имеют связи с другими характеристиками.

Посчитаем медиану и среднее для категорий по операции и возрасту:

```python
h3.groupby(['surgery', 'Age']).mean()
```

```python
h3.groupby(['surgery', 'Age']).median()
```

Заполним пропуски по категориям. Сигма(СКО) у пульса гораздо выше --> заполним пропуски медианой. У температуры отклонение маленькое --> поставим среднее:

```python
h3['pulse'].fillna(h3.groupby(['surgery', 'Age'])['pulse'].transform('median'), inplace=True)
```

```python
h3['rectal temperature'].fillna(h3.groupby(['surgery', 'Age'])['rectal temperature'].transform('mean'), inplace=True)
```

```python
h3['pulse'].isnull().sum()
```

### Проделаем то же самое для respiratory rate, temperature of extremities и Pain

```python
q1 = h3['respiratory rate'].quantile(0.25)
q3 = h3['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers_r = h3[h3['respiratory rate'].between(lower_bound, upper_bound, inclusive=True)].sort_values('respiratory rate')
remove_outliers_r
```

```python
pd.concat([h3, remove_outliers_r]).drop_duplicates(keep=False).dropna()
```
