```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 17,6
```

```python
df = pd.read_csv('2019.csv')
```

```python
df.info()
df.set_index(['Overall rank'], inplace = True)
```

```python
df.columns = df.columns.str.replace(' ','_')
df
```

```python
plt.plot(df.Score, df.GDP_per_capita)
# plt.title('Связь ВВП на душу населения и общего счёта') <--- выдаёт ошибку.
plt.xlabel('Score')
plt.ylabel('GPD per capita')
```

### Вывод: конечно в целом график возрастает, однако явно видно, что общий счёт счастья далеко не всегда зависит от ВВП, т.е. у многих стран он больше, чем у более счастливых жителей, и наоборот.
