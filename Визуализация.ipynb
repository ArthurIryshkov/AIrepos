```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 17,6
```

```python
df = pd.read_csv('2019.csv')
```

```python
df.info()
df.set_index(['Overall rank'], inplace = True)
```

```python
df.columns = df.columns.str.replace(' ','_')
df
```

```python
plt.plot(df.Score, df.GDP_per_capita)
# plt.title('Связь ВВП на душу населения и общего счёта') <--- выдаёт ошибку.
plt.xlabel('Score')
plt.ylabel('GPD per capita')
```

### Вывод: конечно в целом график возрастает, однако явно видно, что общий счёт счастья далеко не всегда зависит от ВВП, т.е. у многих стран он больше, чем у более счастливых жителей, и наоборот.

```python
app = pd.read_csv('googleplaystore.csv')
```

```python
app.info()
```

```python
app.Price.unique()
```

```python
app.head(11)
```

```python
app.Category.unique()
```

```python
app100 = app.groupby(['Category'])['Rating'].mean().head(100).sort_values()
```

```python
app100
```

```python
app100.plot(kind='bar', title='Cредний рейтинг приложений по их категориям (топ-100)')
plt.axis([-1, 35, 3.5, 5])
```

### ...видим явный выброс. Какие категории оцениваются выше остальных? Видим, что приложения-мероприятия, образование и дизайн составляют тройку лидеров.

___________________________________________________________________________________________________________

```python
app2 = app.groupby(['Category'])['App'].count().sort_values()
```

```python
app2
```

```python
app2.plot(kind='bar', title='Количество приложений по категориям')
plt.axis([-1, 35, 0, 2000])
```
