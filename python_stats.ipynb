```python
pulse_counts = {}
for p in h3['pulse'].round():
    if p not in pulse_counts:
        pulse_counts[p] = 1
    else:
        pulse_counts[p] += 1
maxp = 0
mode_p = None
for k, v in pulse_counts.items():
    if maxp < v:
        maxp = v
        mode_p = k
print('Значение моды:', mode_p, 'количество встречаемости:', maxp)
# самый частовстречаемый пульс
```

```python
q1 = h3['pulse'].quantile(0.25)
q3 = h3['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = h3[h3['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

```python
pd.concat([h3, remove_outliers]).drop_duplicates(keep=False) # выбросы по пульсу 20 - просто завышенные показатели и NaN.
```

```python
q1 = h3['rectal temperature'].quantile(0.25)
q3 = h3['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers_t = h3[h3['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal temperature')
remove_outliers_t
```
